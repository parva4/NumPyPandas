```python
def inclusive_range(*args):
  numargs=len(args)
  start=0
  step=1
  if numargs<1:
    raise TypeError(f'expected atleast one arg, got {numargs}')
  elif numargs==1:
    stop=args[0]
  elif numargs==2:
    (start,stop)=args
  elif numargs==3:
    (start,stop,step)=args
  else:
    raise TypeError(f'expected atmost 3 args, got {numargs}')
  i=start
  while i<=stop:
    yield i
    i+=step
def main():
  for i in inclusive_range():
    print(i,end=' ')
main()
```

```python
#max out of a list using function
def max(list):
  maxi=0
  x=len(list)
  for i in range(x):
    for j in range(x):
      if list[i]>=list[j]:
        maxi=list[i]
      else:
        break
  print(maxi)
#max([3,2,1])
max((3,2))
```

```python
#dictionary creator
def dictionary_creator(list1,list2):
  i=0
  if len(list1)==len(list2):
    dict={}
    for val in list1:
      dict[val]=list1[i]
      i=i+1
  else:
    print("Error")
  print(dict)
dictionary_creator(['A','B','C'],[1,2,3])
```
