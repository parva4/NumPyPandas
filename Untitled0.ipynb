moving average using rolling

```python
import pandas as pd
dates=pd.date_range(start='2021-01-01',end='2021-01-10',freq='d').to_list()
num=[10,8,4,15,25,2,6,17,8,12]
x=pd.DataFrame({'dates':dates,'numbers':num})
#x.plot.line(x='dates',y='numbers')
x['Moving average']=x.rolling(window=4).mean()
x.plot.line(x='dates',y=['numbers','Moving average']) #plotting MA
```

```python
#center=True
import pandas as pd
dates=pd.date_range(start='2021-01-01',end='2021-01-10',freq='d').to_list()
num=[10,8,4,15,25,2,6,17,8,12]
x=pd.DataFrame({'dates':dates,'numbers':num})
#x.plot.line(x='dates',y='numbers')
x['Rolling average']=x.rolling(window=4,center=True).mean()
#x.plot.line(x='dates',y=['numbers','Moving average']) #plotting MA
x
```

```python
#min periods
import pandas as pd
import numpy as np
dates=pd.date_range(start='2021-01-01',end='2021-01-10',freq='d').to_list()
num=[10,8,4,15,25,2,6,17,8,12]
x=pd.DataFrame({'dates':dates,'numbers':num})
#x.plot.line(x='dates',y='numbers')
x.iloc[5:7,1]=np.NaN
x['Rolling average']=x.rolling(4,min_periods=3).mean()
#x.plot.line(x='dates',y=['numbers','Moving average']) #plotting MA
x
```
