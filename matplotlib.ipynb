```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,label='2x',marker='.',markersize=10,color='black',markeredgecolor='red',linestyle='--')
#to get rid of [<matplotlib.lines.Line2D at 0x7fa638ffdc10>]
plt.title('plotx',fontdict={'fontname':'Times New Roman','fontsize':25})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
